```python
abav = ext_tab['A_LAMBDA/A_V'][np.where(ext_tab['NAME']=='JOHNSON_B')].data
avav = ext_tab['A_LAMBDA/A_V'][np.where(ext_tab['NAME']=='JOHNSON_V')].data
ajav = ext_tab['A_LAMBDA/A_V'][np.where(ext_tab['NAME']=='2MASS_J')].data
akav = ext_tab['A_LAMBDA/A_V'][np.where(ext_tab['NAME']=='2MASS_Ks')].data
    
abav_err = ext_tab['A_LAMBDA/A_V_ERR'][np.where(ext_tab['NAME']=='JOHNSON_B')].data
avav_err = ext_tab['A_LAMBDA/A_V_ERR'][np.where(ext_tab['NAME']=='JOHNSON_V')].data
ajav_err = ext_tab['A_LAMBDA/A_V_ERR'][np.where(ext_tab['NAME']=='2MASS_J')].data
akav_err = ext_tab['A_LAMBDA/A_V_ERR'][np.where(ext_tab['NAME']=='2MASS_Ks')].data
    
noisy_abav = np.random.normal(abav,abav_err,200)
noisy_avav = np.random.normal(avav,avav_err,200)
noisy_ajav = np.random.normal(ajav,ajav_err,200)
noisy_akav = np.random.normal(akav,akav_err,200)

red_rat_dist = (noisy_ajav-noisy_akav)/(noisy_abav-noisy_avav)
ak_ejk_dist = noisy_akav/(noisy_ajav-noisy_akav)
ak_ebv_dist = noisy_akav/(noisy_abav-noisy_avav)

plt.hist(red_rat_dist)
plt.title(r'E(J-K)/E(B-V) {:.3f} +/- {:.3f}'.format(np.median(red_rat_dist),afunc.mad(red_rat_dist)))
print('Med ', np.median(red_rat_dist))
print('MAD ', afunc.mad(red_rat_dist))
plt.show()

plt.hist(ak_ejk_dist)
plt.title(r'A_k/E(J-K) {:.3f} +/- {:.3f}'.format(np.median(ak_ejk_dist),afunc.mad(ak_ejk_dist)))
print('Med ' ,np.median(ak_ejk_dist))
print('MAD ', afunc.mad(ak_ejk_dist))
plt.show()

plt.hist(ak_ebv_dist)
plt.title(r'A_k/E(B-V) {:.3f} +/- {:.3f}'.format(np.median(ak_ebv_dist),afunc.mad(ak_ebv_dist)))
print('Med ',np.median(ak_ebv_dist))
print('MAD ',afunc.mad(ak_ebv_dist))
plt.show()
```

```python
def cardelli(x,rv):
    '''
    Calculate A(Lambda)/A(V) from Cardelli et al. for 0.3 < lambda^-1 microns^-1 < 3.3
    
    Inputs:
    ------
        x: reciprocal wavelength in micron^-1
        rv: R_V value (standard R_V = 3.1)
        
    Output:
    ------
        al_av: A(Lambda)/A(V)
    
    ''' 
        
    if 0.3 <= x < 1.1:
        a = 0.574*(x**1.61)
        b_rv = (-0.527*(x**1.61))/rv
        al_av = a+b_rv
        return al_av
    
    elif 1.1 <= x <= 3.3:
        y = x - 1.82
        a = (1.+0.17699*y-0.50477*(y**2)-0.02427*(y**3)+0.72085*(y**4)+
             0.01979*(y**5)-0.77530*(y**6)+0.32999*(y**7))
        b_rv = (1.41338*y+2.28305*(y**2)+1.07233*(y**3)-5.38434*(y**4)-
                0.62251*(y**5)+5.30260*(y**6)-2.09002*(y**7))/rv
        al_av = a+b_rv
        return al_av
    
    else:
        print('0.3 <= x micron^-1 <3.3 not satisfied')
        return -9999.0
    
########

def cardelli_a(x):
    if 0.3 <= x < 1.1:
        a = 0.574*(x**1.61)
        return a
    
    elif 1.1 <= x <= 3.3:
        y = x - 1.82
        a = (1.+0.17699*y-0.50477*(y**2)-0.02427*(y**3)+0.72085*(y**4)+
             0.01979*(y**5)-0.77530*(y**6)+0.32999*(y**7))
        return a
    
def cardelli_b(x):
    if 0.3 <= x < 1.1:
        b = -0.527*(x**1.61)
        return b
    
    elif 1.1 <= x <= 3.3:
        y = x - 1.82
        b = (1.41338*y+2.28305*(y**2)+1.07233*(y**3)-5.38434*(y**4)-
                0.62251*(y**5)+5.30260*(y**6)-2.09002*(y**7))
        return b
    
def cardelli_alav(x,rv):
    alav = cardelli_a(x)+cardelli_b(x)/rv
    return alav

def cardelli_alebv(x,rv):
    alebv = cardelli_a(x)*rv+cardelli_b(x)/rv
    return alebv

def cardelli_e12ebv(xblue,xred,rv):
    a2a1rv = (cardelli_a(xblue) - cardelli_a(xred))*rv
    b2b1 = cardelli_b(xblue) - cardelli_b(xred)
    e12ebv = a2a1rv + b2b1
    return e12ebv
    
def cardelli_e12av(xblue,xred,rv):
    a2a1 = cardelli_a(xblue) - cardelli_a(xred)
    b2b1rv = (cardelli_b(xblue) - cardelli_b(xred))/rv
    e12av = a2a1 + b2b1rv
    return e12av
```

```python
car_al_av = []
for i in np.linspace(0.3,3.3):
    car_al_av.append(cardelli_alav(i,3.1))

car_band = []
for j in 1/np.array([.365,.445,.551,1.220,1.630,2.190]):
    car_band.append(cardelli_alav(j,3.1))

plt.figure(figsize=[10,10])
plt.plot(np.linspace(0.3,3.3),car_al_av,c='k',label='Cardelli et al. 1989')
plt.scatter(1/np.array([.365,.445,.551,1.220,1.630,2.190]),car_band,c='k',s=100)

filters = ['U','B','V','J','H','K']
for i in range(len(filters)):
    plt.text(1/np.array([.365,.445,.551,1.220,1.630,2.190])[i],car_band[i]-0.075,filters[i])


plt.title(r'Multiband Extinction Relative to $A_V$ (Cardelli et al. 1989)')
plt.xlabel(r'$1/\lambda\,(\mu m^{-1})$')
plt.ylabel(r'$A_\lambda/A_V$')
plt.show()
```
